```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('aqardata_2.csv')
df
```

```python
indexSize = df[ df['size'] > 10000].index
df.drop(indexSize,inplace=True)
indexPrice = df[ df['Pricepm'] > 8000].index
df.drop(indexPrice,inplace=True)
df.drop(['purpose'],axis=1, inplace= True)
df.drop(['sublocation'],axis=1, inplace= True)
df.dropna(inplace=True)
df
```

```python
df= pd.get_dummies(df,drop_first=True)
df = df.reindex(columns = [col for col in df.columns if col != 'Pricepm'] + ['Pricepm'])
# Features
X = df.iloc[:, :-1].values
# Target
y = df.iloc[:,-1].values
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import KFold
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
model = RandomForestRegressor()
kfold = KFold(n_splits=5, shuffle=True, random_state=93)
i=1
MAE_scores = [] 
RMSE_scores = [] 
R2_scores = []

for train_index, test_index in kfold.split(X):
    
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]

    
    model.fit(X_train, y_train)

    
    y_pred = model.predict(X_test)

    print(f"Fold {i}")
    print(f"MAE: {mean_absolute_error(y_test, y_pred):.3f}")
    print(f"RMSE: {np.sqrt(mean_squared_error(y_test, y_pred)):.3f}")
    print(f"R2: {r2_score(y_test, y_pred):.3f}")
    i = i+1
    MAE_scores.append(mean_absolute_error(y_test, y_pred))
    RMSE_scores.append(np.sqrt(mean_squared_error(y_test, y_pred)))
    R2_scores.append(r2_score(y_test, y_pred))


mae_mean = np.mean(MAE_scores)
rmse_mean = np.mean(RMSE_scores)
r2_mean = np.mean(R2_scores)
print("")
print("Mean Scores:")
print(f"MAE: {mae_mean:.3f}")
print(f"RMSE: {rmse_mean:.3f}")
print(f"R2: {r2_mean:.3f}")
```
